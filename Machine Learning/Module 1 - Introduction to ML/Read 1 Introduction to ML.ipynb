## Machine Learning

Machine Learning (ML) is one of the promising approaches to AI. How does machine learning differ from other traditional programming ideas? What makes machine learning so special? We will discuss it now.

Traditional programming or conventional programming relies on the statement:  *Computers do exactly as they are told, no more and no less.* While, machine learning re-defines it as: *Computers can do whatever they learn from data.*

<center>



<p></p>

Fig: Traditional(Conventional) Programming Vs. Machine Learning

</center>

In traditional programming, **you define a model as a set of rules** (program) that operates on the data to give you the desired output. But in machine learning, you provide the data along with the desired output to the machine learning model, and **the model itself learns the set of rules** to produce the desired output. As a result, machine learning models generalize better than the models created using conventional programming.

So far, we have been using the term *learn* quite a lot. So what does *learn* mean? Tom M. Mitchell provided a widely quoted and more formal definition of what learning means in the context of machine learning: 
>**A computer program is said to learn from experience E with respect to some class of tasks T and performance measure P if its performance at tasks in T, as measured by P, improves with experience E.** 

Thus, experience **E** is the past data, task **T** can be anything from regression, classification, clustering, and so on, and performance **P** represents the performance metrics such as MSE, RMSE, log loss, accuracy, etc. As the data increases, the performance of the machine learning model increases as well.


## Components of Machine Learning

Every machine learning problem can be decomposed into four main components:

* Data

* Model

* Learning

* Application





### Data

Let's start with the data. Data can be anything from facts, records, values, text, images, etc. It is the driving force behind the boom of machine learning. Machine learning algorithms require a large amount of data for training. Even for simple problems, you need thousands of samples to train a model, and for complex problems like image and speech recognition, you will need millions of training samples. Even in this day and age of digitization, we still lack the quality data required to train a model. Most of the data available to us are unstructured and contain a lot of unwanted noise. This forces us to spend a considerable amount of time to clean and process the data into a form more suitable for training a machine learning model. So data cleaning and processing is one of the crucial steps in Machine Learning. You will learn about some basic data cleaning techniques in the next chapter. The other thing to keep in mind here is that the data you select for training a model must be a good representative of the data you want the model to generalize to. You cannot train a model on the data of Nepalese and hope for it to predict the life expectancy of Americans.


### Model (Representation)

The second component of machine learning is the model. Models such as Linear regression, Logistic regression, Naive Bayes, Decision trees, etc. for supervised learning and *K-means* for unsupervised learning. Different models are suitable for different types of problems. So you should select the model according to your problem. Selecting a model is equivalent to selecting a particular representation for your data. For example, when you select a linear regression model, you decide to represent the data using a straight line $Y = \beta_0 + \beta_1 X$.


### Learning

You have selected a particular model. The model consists of a number of parameters. Different combinations of these parameters result in different **hypotheses**. These different hypotheses form a common space known as the **hypotheses space**. The main objective of learning is to find the hypothesis that best fits the data. In other words, you try to find the optimal parameters of the model that fits the data best. 

* **Objective Function:** So you need to find the hypothesis that best fits the data. But how do you do this? You define an **objective function**. An objective function provides a measure of how well or how bad a hypothesis fits data. You need to find the parameters that optimize this objective function. *Mean Squared Error (MSE)* and *log loss* are two of the most popular objective function used in linear regression and logistic regression, respectively. 

* **Optimization:** You now have an objective function that you need to optimize to find the best hypothesis (or the optimal parameters). Now how would you optimize the objective function?  Several algorithms can be used to optimize a function. These algorithms search for the optimal parameters in some systematic way. *Gradient Descent* is one of the most popular optimization algorithms used in machine learning and deep learning. 






### Application

Once the machine learning has learned from the data, you can now use it for different applications. Prediction is one of the widely used applications of machine learning. You can use the machine learning model to predict continuous values (regression) or discrete values (classification). You can also use the model to extract various recommendations for movies, songs, products, etc. Similarly, you can use the model to infer various insights, knowledge, and patterns present in the data.


### Example
Now let's understand the components of machine learning better by breaking down a machine learning problem into its components. Consider a classification problem of classifying whether or not a customer buys a particular product. There are 1000 labeled samples of customers. We select *logistic regression* as our model. Now comes the learning part. We define the objective function to be the *log loss function*. We will use *gradient descent* to optimize this objective function. Then we will use *accuracy* to evaluate the performance of our model on new data. Finally, we will apply this machine learning model to predict if a customer will buy the product or not. 

The components of this machine learning problem are:

* **Data:** 1000 labeled samples of customers

* **Model:** Logistic regression

* **Learning**

  * **Objective:** Log-loss
  * **Optimization:** Gradient descent
  * **Evaluation:** Accuracy

* **Application:** Predicting if a customer will buy the product or not

### Supervised Learning

In supervised learning, you pass the samples containing the input features and their desired outputs to the machine learning model for training. The model then learns a mapping from the input features to the desired output. You aim to produce an accurate enough mapping so that the algorithm can predict the output as correctly as possible whenever a new test input is given. As the desired output passed during the training supervises the model to learn the required mapping, this learning method is called **supervised learning.** 

The following figure illustrates the working of supervised learning to predict if a given image is of an apple or not.

First, the images, along with their corresponding labels, are passed to the model for training. Note that these images contain positive as well as negative samples. In our case, images of apples as well as images that are not of apples. Then the model learns a mapping from the input images to their corresponding labels. When a new test data is passed to the model, it needs to predict the corresponding label correctly. In the figure, a new test image of an apple is passed to the model, and it correctly predicts it as an apple.



Now that you have recalled the basic intuition behind supervised learning, let's define it in a more mathematical way. Consider a training set of $n$ samples of the form $\{(x_1, y_1), (x_2, y_2), ....., (x_n, y_n) \}$ where $x_i$ is the feature of $i^{th}$ sample and $y_i$ is it's label. Supervised learning uses this training set to find the mapping $f: X \rightarrow Y$, where $X$ is the input features space and $Y$ is the output space. You can further define a scoring function $g: X \times Y \rightarrow \mathbb{R}$ that can map the possible combinations of elements of $X$ and $Y$ to a particular score. Then the mapping function $f$ can be defined as $f(x) = {\operatorname{arg}}\underset{y}{\operatorname{max}}\ g(x,y)$ that outputs $y$ that has the maximum score.



### Unsupervised Learning

As opposed to supervised learning, in unsupervised learning, you pass the samples containing only the input features to the machine learning model.  The model then extracts different useful features and patterns from the data. As the desired output is no longer supervising the learning process, this learning method is called **unsupervised learning**.

## Reinforcement Learning

 Reinforcement Learning is the type of machine learning where the machine learns on its own using observations gathered from the interaction with its environment to take actions that would maximize the reward or minimize the risk. 

Reinforcement learning problems consist of an agent in an environment. The agent interacts with the environment in discrete time steps. At any particular time step $t$, the agent receives the current state $s_t$ and the reward $r_t$. Note that $r_t$ is the reward obtained by the agent in the time step $t$ due to its action in the time step $t-1$. The agent then selects an action $a_t$ which changes the state of the environment from $s_t$ to $s_{t+1}$. Then the agent receives the state $s_{t+1}$ and the reward $r_{t+1}$ associated with the transition $\{s_t, a_t, s_{t+1} \}$. This is continued for finite or infinite number of time steps. A policy $\pi : A\times S \rightarrow [0,1]$ maps all the possible combinations of states and actions to a probabilistic value between $0$ and $1$ such that for any state-action pair, it gives the probability of selecting the action $a$ given the state $s$ *i.e.* $\pi(a, s) = P(a_t = a| s_t = s)$. The goal of reinforcement learning is to find the optimal policy $\pi_{opt}$ that maximizes the expected cumulative reward.

## Some other types of Machine Learning

The categorization of machine learning into three main types is based on the type of supervision that the model needs from humans. Other than that, we can categorize the machine learning as:

* Model-Based vs. Instance-Based

* Parametric vs Non-parametric

* Probabilistic vs Non-probabilistic

Now let's go through each of them in brief.


### Model-Based vs. Instance-Based 

**Model-Based Learning:** Consider a classification problem. You have a dataset containing features and labels. You train a logistic regression model using this dataset. The logistic regression model learns a decision boundary separating the classes. After that, given any new data point, that model tries to predict a target class using that decision boundary independent of the dataset's original data points.

<p>Model-based learning</p>
</center>

As you can see in this [animation](https://doc.google.com/a/fusemachines.com/uc?export=download&id=1iddcGcMALyIpLMNVAgesmbPu64jRhRQL), the decision boundary separates two different colors of dots. Now, given any unknown black dot, it will classify the unknown dot as green if it is on the right side of the line and blue if it is on the left side of the line. **This method of learning a pattern or insight from the data and using them for prediction is called model-based learning.** 

\

**Instance-Based Learning:** Consider the same classification but this time with K-NN model. The K-NN model doesn't learn any pattern or insight from the data. When a new data point appears, it simply compares the new data point with the original data points in the dataset and predicts the class accordingly.

Let's say you store all the data points on a database/memory (Illustrating on a 2D plot, for example). Now, given any new unknown point (unknown blue cross), the closest/similar data points to that unknown point are filtered from memory, and the label of the most frequent point (triangle in case of above example) is assigned to that unknown. **This method of using the original data points instead of the pattern present in the dataset for prediction is called instance-based learning.**





### Parametric vs Non-parametric

**Parametric Learning:** In parametric learning, a model has a constant and finite number of parameters. Parametric models make stronger assumptions about the data. For example, a simple linear regression model assumes that the data is linear, and a straight line of the form $Y = \beta_0 + \beta_1X$ can fit the data. As a consequence of this assumption, no matter how large the dataset gets, there will be only two parameters: $\beta_0$ and $\beta_1$

Parametric models are computationally faster and perform well if the assumptions hold, but they may perform very poorly if the assumptions are wrong.

\

**Non-parametric Learning:** In non-parametric learning, a model has a variable and possibly an infinite number of parameters. Non-parametric models do not make assumptions about the data and contain a flexible number of parameters. For example, in a decision tree, the number of parameters *, i.e.,* the tree size is not fixed and increases exponentially with the size of the dataset.

Non-Parametric models are computationally slower but generally perform well as they don't make any strong assumptions about the data.




### Probabilistic vs Non-probabilistic

**Probabilistic models:** Considering a Naive Bayes classifier, it calculates the probability of the sample belonging to each class given its features. For a given input, probabilistic models are the ones that provide a probability for each class as the output. You can use the `predict_proba` method from sklearn to print these probabilities. Then usually, the class with the highest probability is predicted. Naive Bayes and Logistic Regression are two of the most popular probabilistic models. 

\

**Non-probabilistic models:** The models that directly output the predicted class instead of the probabilities are known as non-probabilistic models, such as Decision Trees, SVM, and K-NN, etc. Sklearn allows you to call the `predict_proba` for non-probabilistic models as well. However, unlike in probabilistic models, the predictions are not derived from these probabilistic values.


## Benefits and Limitations of Machine Learning


### Benefits

**Predictive Prowess:** Hundred years ago, if you said you could predict the future, everyone would think that you're either a sorcerer or an insane person. But now, there are models that can predict the future so accurately that they'll put sorcerors in shame. All thanks to machine learning. Machine learning has been used to predict almost everything from weather, diseases, and earthquakes to your future appearance, Netflix recommendations, and face tagging. Without machine learning, all of these applications would have been a subject for fantasy and fiction.

**Extract insights and patterns from data:** In this day and age of digitization, we have a large amount of data available. Manually analyzing this huge amount of data and extracting useful information from it is impossible. This is where machine learning comes in. It can extract useful insights and patterns from this huge amount of data. Moreover, machine learning models can detect the insights and patterns which are beyond the comprehension of a human mind and could never be detected manually. Isn't that amazing? 

**Automation:** Machine learning has provided a great deal of automation for our complex repetitive tasks. It has been widely used for tasks such as fraud detection, disease diagnosis, data entry, chatbots, industrial automation, etc. Everyday innovations are being made in the ambitious efforts of self-driving cars, AI-powered personal assistants, social humanoid robots, etc., that will step up the game of automation in human history. This larger number of automation will allow humans to focus on more critical tasks.


**Continuous Improvement:** Machine learning models improve continuously over time.  They don't need regular human intervention like traditional programs to perform better. All they need is data. They will automatically learn and adapt to newer data and act accordingly.


### Limitations

**Lack of Quality Data:** You have already learned that data is the most important component of machine learning. As the saying goes, "Too much data is never enough" we are always short of data. Machine learning models require a huge amount of data for training. Moreover, most of the data available are unstructured and contain a lot of noise. Machine learning models cannot learn anything from this noisy and unstructured data. Also, the data you select for training a model must be a good representative of the data you want the model to generalize to. So there is a great challenge to acquire and process the data that can be used to train a well-performing model. 

**Time and Resources:** Training a model to perform complex tasks such as image and speech recognition takes a lot of time. You can use advanced resources like GPU and TPU, but they are either too expensive or not always available. 

**Lack of Interpretation:** As stated earlier, machine learning models can detect the insights and patterns beyond the comprehension of a human mind. This makes the working of machine learning models less interpretable and intuitive. This is more evident in computer vision tasks where a model picks up weird pixels, which seem like a mere noise to human eyes as features.
